# Extra work

```python
from pandas_datareader import data
import datetime as dt
import matplotlib.pyplot as plt
import numpy as np
import yfinance as yf
```

```python
start = dt.datetime(2022,9,6)
end = dt.date.today()

yf.pdr_override()
price_AAPL = data.DataReader('AAPL', start, end)
```

```python
price_AAPL.head()
```

```python
price_AAPL['daily_differnce'] = price_AAPL['Adj Close'].diff()
price_AAPL.head()
```

```python
price_AAPL['signal'] = np.where(price_AAPL['daily_differnce']>0, 1, 0)
price_AAPL.head()
```

```python
price_AAPL['possition'] = price_AAPL['signal'].diff()
price_AAPL.head()
```

```python
price_AAPL[price_AAPL['possition']==1.0].index
```

```python
price_AAPL.loc[price_AAPL['possition']==1.0, 'Adj Close']
```

```python
plt.figure(figsize = (10,6))

plt.plot(price_AAPL['Adj Close'], c = 'r', lw = 2)

plt.plot(price_AAPL[price_AAPL['possition']==1.0].index,
        price_AAPL.loc[price_AAPL['possition']==1.0, 'Adj Close'], 'o', c = 'g', markersize = 5
        )

plt.plot(price_AAPL[price_AAPL['possition']== -1.0].index,
        price_AAPL.loc[price_AAPL['possition']== -1.0, 'Adj Close'], 'x', c = 'k', markersize = 5
        )

plt.grid()
plt.show()
```

## 4. World Bank

### 4.1 Imports

```python
from pandas_datareader import wb
import datetime as dt
```

```python
inf = wb.download(country='all', indicator='FP.CPI.TOTL.ZG', start = 1960, end = 2023)
```

```python
inf.head()
```

```python
inf.unstack().head()
```

```python
inf2 = inf.unstack()
```

```python
inf2.columns = np.arange(1960, 2023)
```

```python
inf2.head()
```

```python
inf3 = inf2.T
inf3.head()
```

```python
inf3['Iran, Islamic Rep.'].plot(figsize=(10,6), lw = 2)

plt.grid()
plt.xlabel('Years', fontsize = 10, fontname = 'Times New Roman')
plt.ylabel('percent', fontsize = 10, fontname = 'Times New Roman')
plt.title("Iran's Inflation", fontsize = 14, fontname = 'Times New Roman')


plt.legend(fontsize = 10)
plt.show()
```

```python
inf3[['Iran, Islamic Rep.', 'China']]#.dropna()
```

```python
inf3[['Iran, Islamic Rep.', 'China']].dropna().plot(figsize=(10,6), lw = 2)

plt.grid()
plt.xlabel('Years', fontsize = 10, fontname = 'Times New Roman')
plt.ylabel('percent', fontsize = 10, fontname = 'Times New Roman')
plt.title("Iran and China Inflation", fontsize = 14, fontname = 'Times New Roman')


plt.legend(fontsize = 10)
plt.show()
```

### 4.2 Second data

```python
exp = wb.download(country='all', indicator='TX.VAL.MRCH.XD.WD', start = 1960, end = 2023)
```

```python
exp2 = exp.unstack()
exp2.columns = np.arange(1960,2023)
exp3 = exp2.T
exp3.head()
```

```python
exp3[['Iran, Islamic Rep.', 'Iraq']].dropna().plot(figsize=(10,6), lw = 2)

plt.grid()
plt.xlabel('Years', fontsize = 10, fontname = 'Times New Roman')
plt.ylabel('percent', fontsize = 10, fontname = 'Times New Roman')
plt.title("Iran's Export index", fontsize = 14, fontname = 'Times New Roman')


plt.legend(fontsize = 10)
plt.show()
```

### 4.3 concat

```python
data = pd.concat([exp3['Iran, Islamic Rep.'], exp3['Iraq']], axis = 1)
data.columns = ['Iran', 'Iraq']
data.dropna(inplace=True)
data
```
