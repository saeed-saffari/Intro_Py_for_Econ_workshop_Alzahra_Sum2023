### Task 2.3: find maximum infection rate for all of the countries. 

```python
country = list(corona_dataset_aggegated.index)
```

```python
max_infection = []

for c in country:
    max_infection.append(corona_dataset_aggegated.loc[c].diff().max())
    
corona_dataset_aggegated['max_infection_rate'] = max_infection
```

```python
corona_dataset_aggegated.head()
```

### Task 2.4: create a new dataframe with only needed column 

```python
corona_2 = pd.DataFrame(corona_dataset_aggegated['max_infection_rate'])
```

```python
corona_2.head()
```

### Task 5.1: Plotting GDP vs maximum Infection rate

```python
x = data['Logged GDP per capita']
y = data['max_infection_rate']

sns.scatterplot(x = x, y = np.log(y))
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.regplot(x = x, y = np.log(y))
plt.title('GDP per Capita VS Max Infection Rate', fontsize = 15, fontweight = 3, fontname="Times New Roman")
plt.grid()
plt.show()
```
